```python
import random
import csv
import math

npoints = 15 #number of the coordinates to be created 


data = [(int(random.random()*10), int(random.random()*10)) for _ in range(npoints)]
start_point = [(0,0)]

# with open(r'C:\Users\kamil\Desktop\CIty\Programming\\COURSEWORK.csv', 'w') as file:
#     writer = csv.writer(file)
#     writer.writerow(data)
#     writer.writerow(start_point)

# with open(r'C:\Users\kamil\Desktop\CIty\Programming\\COURSEWORK.csv', 'r') as file:
#     reader = csv.reader(file)
#     for row in reader:
#         print(row)

print(data)
```

```python
distance_list = []
q = start_point[0]
path= [q]
pathpoint = 1
TotalDst=0
ALL=[]
print("that's what the data looks like", data)
while npoints >= 1:
#     print("-----")
    if len(path) == 1:
        path = [q]
        for x in data:
            p = x
            distance = math.dist(p, q)
            print(p, q)
            distance_list.append(distance)
            ALL.append(distance)
            ALL.sort()
            if len(distance_list) == npoints:
                print("-----")
                print("print the entire list", distance_list)
                nearest = min(distance_list)
                d = distance_list.index(nearest)
                print("print the nearest distance coordinate :",nearest, "and its index number :", d)
                nearest_data = data[d]
                path.append(nearest_data)
                TotalDst= TotalDst + nearest
                data.pop(d)
                distance_list.pop(d)
                npoints= npoints -1
                for check in distance_list:
                    z = nearest
                    if check == z:
                        nearest = min(distance_list)
                        d = distance_list.index(nearest)
                        nearest_data = data[d]
                        path.append(nearest_data)
                        TotalDst= TotalDst + nearest
                        data.pop(d)
                        distance_list.pop(d)
                        npoints= npoints - 1
                        pathpoint = pathpoint + 1
                        print("duplicate")
                        print("-----")
                        print("update the path by duplicate",path)
                        break
                print("current path",path)
                print("-----")
                print("current data",data)
                distance_list = [] 
                print("end loop nearest to zero")
                print("-----")
                break
    if len(path) > 1:
        q= path[pathpoint]
        for x in data:
            p = x
            distance = math.dist(p, q)
            print(p, q)
            distance_list.append(distance)
            if len(distance_list) == npoints:
                print("print the entire list that left", distance_list)
                nearest = min(distance_list)
                d = distance_list.index(nearest)
                print("print the nearest distance coordinate out of the list :", nearest, "and its index number :", d)
                nearest_data = data[d]
                path.append(nearest_data)
                TotalDst= TotalDst + nearest
                data.pop(d)
                distance_list.pop(d)
                npoints= npoints - 1
                pathpoint = pathpoint + 1
                for check in distance_list:
                    z = nearest
                    if check == z:
                        nearest = min(distance_list)
                        d = distance_list.index(nearest)
                        nearest_data = data[d]
                        path.append(nearest_data)
                        TotalDst= TotalDst + nearest
                        data.pop(d)
                        distance_list.pop(d)
                        npoints= npoints - 1
                        pathpoint = pathpoint + 1
                        print("duplicate")
                        print("-----")
                        print("update the path by duplicate",path)
                        break
                distance_list = []
                print("current path",path)
                print("-----")
                print("current data",data)
                print ("next nearest point ends loop")
                break
print("-----")
print("the total path in the sort maner",path)            
print("The total distance from Depo through every single Drop off point back to Depo is ",TotalDst+(math.dist((path[-1]), (0,0))))
print("end",data)
```
